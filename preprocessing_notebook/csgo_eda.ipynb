# How many events, matches, teams, and players in each country?

```python
events_country = pd.DataFrame(players.groupby('country')[['event_id', 'match_id', 'player_id', 'team']].nunique()).reset_index()
events_country.columns = ['country', 'total_events', 'total_matches', 'total_players', 'total_teams']
events_country.sort_values('total_events', ascending=False, inplace=True)
```

```python
events_country['total_matches/event'] = np.round(events_country['total_matches']/events_country['total_events'],0)
events_country['total_teams/event'] = np.round(events_country['total_teams']/events_country['total_events'],0)
events_country['total_players/event'] = np.round(events_country['total_players']/events_country['total_events'],0)
```

# Top 10 countries that holds the most events

```python
players['year_match'] = pd.DatetimeIndex(players['date']).year
```

```python
yearly_events = pd.DataFrame(players.groupby(['year_match', 'country'])[['event_id']].nunique()).reset_index()
yearly_events.columns = ['year_match', 'country', 'total_events']
yearly_events.sort_values(['total_events'], ascending=True, inplace=True)
```

```python
#yearly_events.to_csv('yearly_events.csv')
```

```python
top_10_event_country = pd.DataFrame(yearly_events.groupby('country').aggregate({'total_events':'sum'})).reset_index().sort_values('total_events', ascending=False).head(10)
```

```python
top_10_event_country
```

```python
top_10_country = np.array(top_10_event_country.country)
```

# Most picked map

```python
picks
```

```python
map_picks = pd.DataFrame(pd.get_dummies(picks['left_over']).sum().reset_index())
map_picks.columns = ['map_names', 'total_picks']
map_picks.sort_values('total_picks', ascending=False, inplace=True)
```

```python
# First removed
map_first_removed_t1 = pd.DataFrame(pd.get_dummies(picks['t1_removed_1']).sum().reset_index())
map_first_removed_t1.columns = ['map_names', 'total_first_removed_t1']
map_first_removed_t1.sort_values('total_first_removed_t1', ascending=False, inplace=True)

map_first_removed_t2 = pd.DataFrame(pd.get_dummies(picks['t2_removed_1']).sum().reset_index())
map_first_removed_t2.columns = ['map_names', 'total_first_removed_t2']
map_first_removed_t2.sort_values('total_first_removed_t2', ascending=False, inplace=True)

map_first_removed = map_first_removed_t1.merge(map_first_removed_t2, on='map_names', how='inner')
map_first_removed["total_first_removed"] = map_first_removed['total_first_removed_t1']+map_first_removed['total_first_removed_t2']
map_first_removed.drop(['total_first_removed_t1', 'total_first_removed_t2'], inplace=True, axis=1)
```

```python
# Second removed
map_second_removed_t1 = pd.DataFrame(pd.get_dummies(picks['t1_removed_2']).sum().reset_index())
map_second_removed_t1.columns = ['map_names', 'total_second_removed_t1']
map_second_removed_t1.sort_values('total_second_removed_t1', ascending=False, inplace=True)

map_second_removed_t2 = pd.DataFrame(pd.get_dummies(picks['t2_removed_2']).sum().reset_index())
map_second_removed_t2.columns = ['map_names', 'total_second_removed_t2']
map_second_removed_t2.sort_values('total_second_removed_t2', ascending=False, inplace=True)

map_second_removed = map_second_removed_t1.merge(map_second_removed_t2, on='map_names', how='inner')
map_second_removed["total_second_removed"] = map_second_removed['total_second_removed_t1']+map_second_removed['total_second_removed_t2']
map_second_removed.drop(['total_second_removed_t1', 'total_second_removed_t2'], inplace=True, axis=1)
```

```python
# Third removed
map_third_removed_t1 = pd.DataFrame(pd.get_dummies(picks['t1_removed_3']).sum().reset_index())
map_third_removed_t1.columns = ['map_names', 'total_third_removed_t1']
map_third_removed_t1.sort_values('total_third_removed_t1', ascending=False, inplace=True)

map_third_removed_t2 = pd.DataFrame(pd.get_dummies(picks['t2_removed_3']).sum().reset_index())
map_third_removed_t2.columns = ['map_names', 'total_third_removed_t2']
map_third_removed_t2.sort_values('total_third_removed_t2', ascending=False, inplace=True)

map_third_removed = map_third_removed_t1.merge(map_third_removed_t2, on='map_names', how='inner')
map_third_removed["total_third_removed"] = map_third_removed['total_third_removed_t1']+map_third_removed['total_third_removed_t2']
map_third_removed.drop(['total_third_removed_t1', 'total_third_removed_t2'], inplace=True, axis=1)
```

```python
most_picked_removed_map = pd.merge(pd.merge(pd.merge(map_picks.head(9), map_first_removed, on='map_names'), map_second_removed, on='map_names'), map_third_removed, on='map_names')
```

```python
most_picked_removed_map
```

```python
most_picked_removed_map = most_picked_removed_map.melt(id_vars=['map_names'], value_vars=['total_picks', 'total_first_removed', 'total_second_removed', 'total_third_removed'])
```

```python
most_picked_map = most_picked_removed_map.loc[most_picked_removed_map["variable"]=="total_picks"]
```

```python
most_removed_map = most_picked_removed_map.loc[most_picked_removed_map["variable"].isin(['total_first_removed', 'total_second_removed', 'total_third_removed'])]
```

```python
most_picked_map.to_csv('most_picked_map.csv')
most_removed_map.to_csv('most_removed_map.csv')
```

# Teams with most victories

```python
results['winning_team']=0
results['losing_team']=0
```

```python
for index, value in results.match_winner.iteritems():
  if value==1:
    results['winning_team'].iloc[index] = results['team_1'].iloc[index]
    results['losing_team'].iloc[index] = results['team_2'].iloc[index]
  elif value==2:
    results['winning_team'].iloc[index] = results['team_2'].iloc[index]
    results['losing_team'].iloc[index] = results['team_1'].iloc[index]
```

```python
all_time_result = results[['date', 'event_id', 'match_id', '_map', 'team_1', 'team_2', 'rank_1', 'rank_2', 'winning_team', 'losing_team']]
```

```python
all_time_result
```

```python
team_victories = pd.DataFrame(all_time_result.groupby('winning_team')['team_1'].count().reset_index())
team_victories.columns = ['team_name', 'total_victories']
team_victories.sort_values('total_victories', ascending=False, inplace=True)

team_losses = pd.DataFrame(all_time_result.groupby('losing_team')['team_1'].count().reset_index())
team_losses.columns = ['team_name', 'total_losses']
team_losses.sort_values('total_losses', ascending=False, inplace=True)
```

```python
team_victories
```

```python
team_losses
```

# TOP 10 Victorious Teams

```python
top_10_victorious_teams = np.array(team_victories.team_name.head(10))
```

```python
top_10_all_time_teams = all_time_result.loc[all_time_result.winning_team.isin(top_10_victorious_teams)][["date","winning_team", "_map"]]
top_10_all_time_teams["year_match"] = pd.to_datetime(top_10_all_time_teams.date).dt.to_period('Y')
```

```python
top_10_all_time_teams
```

```python
top_10_winning_teams = top_10_all_time_teams.groupby(["year_match", "winning_team"]).aggregate({"winning_team":"count"})
```

```python
top_10_winning_teams.columns = ['total_wins']
```

```python
top_10_winning_teams.reset_index()
```

```python
top_10_winning_teams.to_csv("top_10_winning_teams.csv")
```
